```python
import requests
from bs4 import BeautifulSoup
# Make a request to the URL
response = requests.get('https://www.ukuni.net/universities')
# Parse the HTML of the page
soup = BeautifulSoup(response.text, 'html.parser')
# Find all the links on the page
links = soup.find_all('a')
Name_0=[]
# Iterate through the links
for link in links:
  # Check if the link has a hreflang attribute with the value "en"
  if link.has_attr('hreflang') and link['hreflang'] == 'en':
    # Print the text of the link
    # print(link.text)
    Name_0.append(link.text)
```

```python
Name_0 #Do trang này viết 2 lần nên nó bị lặp, ta sẽ lấy 1 tên thôi
```

```python
Name=[]
for i in range(0,100,2):
  Name.append(Name_0[i])
```

```python
len(Name) # lấy thông tin được 50 trường
```

```python
# Find all of the div elements with the desired class
divs = soup.find_all('div', class_='university-right-conbox-right-botconbg-innbotbg')
# Create an empty list to store the numbers
numbers = []
# Loop through the div elements
for div in divs:
    # Extract the number from the div element
    number = div.text.strip()
    # Add the number to the list
    numbers.append(number)
# Print the list of numbers
print(numbers)
```

```python
len(numbers)
```

```python
# ở biến numbers gồm lần lượt các giá trị: Satisfaction, Ave Tuition Fees, Times Ranking,International
#Ta lọc dần ra là đc
```

```python
Satisfaction=[]
for i in range(0,200,4):# Lấy các giá trị ở vị trí 0,4,8....
  Satisfaction.append(numbers[i])
Satisfaction 
#len(Satisfaction)
```

```python
Fees=[]
for i in range(1,200,4):
  Fees.append(numbers[i])
Fees 
#len(Fees)
```

```python
Ranking=[]
for i in range(2,200,4):
  Ranking.append(numbers[i])
Ranking 
#len(Ranking)
```

```python
International=[]
for i in range(3,200,4):
  International.append(numbers[i])
International 
#len(International)
```

```python
import pandas as pd
data=pd.DataFrame(list(zip(Name,Satisfaction,Fees,Ranking,International)),
               columns =['Name','Satisfaction','Fees','Ranking','International'])
```

```python
data.head()
```

```python
data.info() #Bảng data đang dùng có dạng chữ, ta cần chuyển sang dạng số
```

```python
#Chuyển data sang dạng số
data['Satisfaction'] = data['Satisfaction'].str.replace('%', '').astype(float)
data['Fees'] = data['Fees'].str.replace(',', '').astype(float)
data['Ranking'] = data['Ranking'].str.extract('(\d+)').astype(float)
data['International'] = data['International'].str.replace('%', '').astype(float)
```

```python
data.head() 
```

```python
X = data.iloc[:, 1:].values # Dữ liệu cho training. phân loại 
from sklearn.preprocessing import StandardScaler
scaler = StandardScaler()
scaled_features = scaler.fit_transform(X) # Transform data để phù hợp với đầu vào của các hàm tính toán dưới đây
#scaled_features
```

```python
from kneed import KneeLocator
from sklearn.metrics import silhouette_score
from sklearn.cluster import KMeans
```

```python
#1. Dùng elbow với giá trị SSE
```

```python
kmeans_kwargs = {}
sse = [] #Tính toán giá trị SSE cho mỗi giá trị của k
for k in range(1, 20):
 kmeans = KMeans(n_clusters=k, **kmeans_kwargs)
 kmeans.fit(scaled_features)
 sse.append(kmeans.inertia_)
```

```python
#import thư viện
import numpy as np, pandas as pd
import matplotlib.pyplot as plt #Thư viện cho vẽ hình
import seaborn as sns; sns.set()  
plt.rcParams.update({'figure.figsize':(7,3), 'figure.dpi':120}) #Cài đặt kích thước cho toàn bộ các hình trong file notebook này giống nhau
```

```python
plt.rcParams.update({'figure.figsize':(6,3), 'figure.dpi':120})
plt.plot(range(1, 20), sse)
plt.xticks(range(1, 20))
plt.xlabel("Number of Clusters")
plt.ylabel("SSE")
plt.show()
```

```python
kl = KneeLocator(range(1, 20), sse, curve="convex", direction="decreasing") #Tính toán, hiển thị ra giá trị elbow
kl.elbow
```

```python
#2. Dùng silhouette coefficients
```

```python
silhouette_coefficients = []
for k in range(2, 20):
 kmeans = KMeans(n_clusters=k, **kmeans_kwargs)
 kmeans.fit(scaled_features)
 score = silhouette_score(scaled_features, kmeans.labels_)
 silhouette_coefficients.append(score)
```

```python
plt.plot(range(2, 20), silhouette_coefficients)
plt.xticks(range(2, 20))
plt.xlabel("Number of Clusters")
plt.ylabel("Silhouette Coefficient")
plt.show()
```

```python
#2. Dùng elbow với giá trị wcss
```

```python
wcss = []
for i in range(1, 20):
    kmeans = KMeans(n_clusters = i, init = 'k-means++', random_state = 42)
    kmeans.fit(X)
    wcss.append(kmeans.inertia_)
plt.plot(range(1, 20), wcss)
plt.title('The Elbow Method')
plt.xlabel('Number of clusters')
plt.ylabel('WCSS')
plt.show()
```

```python
# Chọn số nhóm n = 3
```

```python
kmeans = KMeans(n_clusters = 3)
y_pred = kmeans.fit_predict(X)
```

```python
data['Class']=y_pred  #Add thêm cột class vào bảng data
```
